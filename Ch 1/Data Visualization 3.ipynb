```python
from bokeh.charts import Bar, show
import pandas as pd

cyls = [11, 7, 14]
labels = ["4", "6", "8"]
cyl_df = pd.DataFrame({
    "cyl": cyls,
    "label": labels
})

bar = Bar(cyl_df, values = "cyl", label = "label")
show(bar)
```
